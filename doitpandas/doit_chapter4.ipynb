### matplotilb 으로 그래프 그리기  

1. 전체 그래프가 위치할 기본 틀을 만든다  
2. 그래프를 그려 넣을 그래프 격자를 만든다  
3. 격자에 그래프를 하나씩 추가한다. 격자는 왼쪽에서 오른쪽 방향으로 추가된다  
4. 만약 격자의 첫번째 행이 꽉 차면 두 번째 행에 그래프를 그려 넣는다

```python
#한 번에 4개의 그래프 그리기

# 데이터 저장
dataset_1 = anscombe[anscombe['dataset'] == 'I']
dataset_2 = anscombe[anscombe['dataset'] == 'II']
dataset_3 = anscombe[anscombe['dataset'] == 'III']
dataset_4 = anscombe[anscombe['dataset'] == 'IV']
```

```python
#기본 틀 제작

fig = plt.figure() # 빈 틀 생성

#격자 추가
axes1 = fig.add_subplot(2,2,1)
axes2 = fig.add_subplot(2,2,2)
axes3 = fig.add_subplot(2,2,3)
axes4 = fig.add_subplot(2,2,4)

#격자에 데이터 전달
axes1.plot(dataset_1['x'], dataset_1['y'], 'o')
axes2.plot(dataset_2['x'], dataset_2['y'], 'o')
axes3.plot(dataset_3['x'], dataset_3['y'], 'o')
axes4.plot(dataset_4['x'], dataset_4['y'], 'o')

#격자에 제목 추가 - set_title 메서드
axes1.set_title("dataset 1")
axes2.set_title("dataset 2")
axes3.set_title("dataset 3")
axes4.set_title("dataset 4")

#기본 틀에 제목 추가 - suptitle 메서드
fig.suptitle("Anscombe Data")

#레이아웃 조절 - tight_layout 메서드
fig.tight_layout()

#그래프 확인
fig
```

seaborn 라이브러리에서 tips 라는 데이터 집합을 사용하여 그래프 그리기

```python
#tips 데이터를 변수 tips 에 저장
tips = sns.load_dataset("tips")
print(tips.head())
print(type(tips))
```

```python
#히스토그램 그리기

#기본 틀 마련
fig = plt.figure()

#그래프 격자 구성
axes1 = fig.add_subplot(1,1,1) #(행개수, 열개수, 번호)

#hist 메서드로 히스토그램 만들기
axes1.hist(tips['total_bill'], bins = 10) #bins 로 x축 간격 조정
axes1.set_title('Histogram of Total Bill')
axes1.set_xlabel('Frequency')
axes1.set_ylabel('Total Bill')
```

```python
#산점도 그래프 그리기 - 산점도 그래프 : 변수 2개를 사용해서 만드는 그래프(이변량 그래프)

scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1,1,1)

#scatter 메서드에 total_bill, tips 열 전달
axes1.scatter(tips['total_bill'],tips['tip'])

axes1.set_title('Scatterplot of Total Bill vs Tip')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')
```

```python
#박스 그래프 그리기 - 이산형 변수와 연속형 변수를 함께 사용

boxplot = plt.figure()
axes1 = boxplot.add_subplot(1,1,1)

#boxplot 메서드
axes1.boxplot([tips[tips['sex'] == 'Female']['tip'],
               tips[tips['sex'] == 'Male']['tip']],
               labels = ['Female', 'Male'])

axes1.set_title('Boxplot of Tips by Sex')
axes1.set_xlabel('Sex')
axes1.set_ylabel('Tip')
```

#### 다변량 데이터로 다변량 그래프 그리기 - 산점도 그래프

```python
#Female 0 반환, Male 1 반환하는 함수

def recode_sex(sex):
    if sex == 'Female':
        return 0
    else:
        return 1

#recode_sex 메서드의 반환값은 데이터프레임에 추가
tips['sex_color'] = tips['sex'].apply(recode_sex)
```

```python
scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1,1,1)
axes1.scatter(
x=tips['total_bill'],
y=tips['tip'],
s=tips['size']*10, #테이블당 인원수
c=tips['sex_color'], #성별의 치환값
alpha=0.5)

axes1.set_title('Total Bill vs Tip Colored by Sex and Sized by Size')
axes1.set_xlabel('Sex')
axes1.set_ylabel('Tip')
```

### seaborn 라이브러리로 그래프 그리기

```python
import seaborn as sns

tips  = sns.load_dataset("tips")

#히스토그램 그리기 - subplots, distplot 메서드
ax = plt.subplots() #기본 틀 제작
ax = sns.distplot(tips['total_bill']) #distplot 메서드에 열 데이터 전달
ax.set_title('Total Bill Histogram with Density Plot')

#경고 메시지는 메서드의 사용 방법이 바뀌었음을 알림
```

```python
ax = plt.subplots()
ax = sns.distplot(tips['total_bill'], kde = False) #밀집도 그래프 제외
ax.set_title('Total Bill Histogram')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Frequency')

#hist = False 로 지정하면 밀집도 그래프만 나온다
```

```python
ax = plt.subplots()
ax = sns.distplot(tips['total_bill'], rug = True) #rug 그래프 추가
ax.set_title('Total Bill Histogram with Density and Rug Plot')
ax.set_xlabel('Total Bill')
ax.set_ylabel('')
##y축에 왜 Density가 뜨는가? 강제로 지우긴함
```

```python
#count 그래프 그리기

ax = plt.subplots()
ax = sns.countplot('day', data = tips) #countplot 메서드
ax.set_title('Count of days')
ax.set_xlabel('Day of the Week')
ax.set_ylabel('Frequency')
```

```python
#seaborn로 산점도 그래프와 회귀선 그리기 - regplot 메서드
ax = plt.subplots()
ax = sns.regplot(x='total_bill', y='tip', data=tips, fit_reg = False) #회귀선 사용 여부
ax.set_title('Scatterplot of Total Bill and Tip')
ax.set_xlabel('Total Bil')
ax.set_ylabel('Tip')
```

```python
#산점도 그래프와 히스토그램 한번에 그리기 - jointplot 메서드
joint = sns.jointplot(x='total_bill', y='tip', data = tips)
joint.set_axis_labels(xlabel='Total Bill', ylabel='Tip')
joint.fig.suptitle('Joint Plot of Total Bill and Tip', fontsize=10, y=1.03)
```

```python
#육각 그래프 그리기 - jointplot 의 인자값 kind = hex
hexbin = sns.jointplot(x='total_bill', y='tip', data = tips, kind="hex")
hexbin.set_axis_labels(xlabel='Total Bill', ylabel='Tip')
hexbin.fig.suptitle('Joint Plot of Total Bill and Tip', fontsize=10, y=1.03)
```

```python
#이차원 밀집도 그리기 - kdeplot 메서드
ax = plt.subplots()
ax = sns.kdeplot(data=tips['total_bill'],
                data2=tips['tip'],
                shade=True) #그래프에 음영 효과

ax.set_title('Kernel Density Plot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```

```python
#바 그래프 그리기 - barplot 메서드
ax = plt.subplots()
ax = sns.barplot(x='time', y='total_bill', data=tips)
ax.set_title('Bar plot of average total bill for time of day')
ax.set_xlabel('Time of day')
ax.set_ylabel('Average total bill')
```

```python
#박스 그래프 그리기 - boxplot
ax = plt.subplots()
ax = sns.boxplot(x='time', y='total_bill', data=tips)
ax.set_title('Boxplot of total bill by time of day')
ax.set_xlabel('Time of day')
ax.set_ylabel('Average total bill')
```

```python
#바이올린 그래프 - violinplot 메서드
ax = plt.subplots()
ax = sns.violinplot(x='time', y='total_bill', data=tips , hue='sex', split=True) #hue 인자로 색상추가
ax.set_title('Violinplot of total bill by time of day')
ax.set_xlabel('Time of day')
ax.set_ylabel('Average total bill')
```

```python
#관계 그래프 그리기 - pairplot 메서드

fig = sns.pairplot(tips, hue='sex')
```

```python
#map_upper, map_lower 메서드
pair_grid = sns.PairGrid(tips)
pair_grid = pair_grid.map_upper(sns.regplot)
pair_grid = pair_grid.map_lower(sns.kdeplot)
pair_grid = pair_grid.map_diag(sns.distplot, rug=True)
plt.show()

## 가운데 그래프가 안그려짐
```

```python
#산점도 그래프의 크기와 모양 조절하기 -kew 인자에 딕셔너리 형태로 전달
scatter = sns.lmplot(x='total_bill', y='tip', data=tips, fit_reg=False, hue='sex', scatter_kws={'s': tips['size']})
##스칼라 크기 문제로 인한 Error
```

```python
#lmplot으로 그래프 한 번에 그리기 - col 인자
anscombe_plot = sns.lmplot(x='x', y='y', data=anscombe, fit_reg=False, col='dataset', col_wrap=2)
```

```python
#FacetGrid 클래스로 그룹별 그래프 그리기

facet = sns.FacetGrid(tips, col='time')
facet.map(sns.distplot, 'total_bill', rug=True)
```

```python
#facet으로 다변량 그래프 그리기
facet = sns.FacetGrid(tips, col='day', hue='sex')
facet = facet.map(plt.scatter, 'total_bill', 'tip')
facet = facet.add_legend()
```

### seaborn 라이브러리로 그래프 스타일 정하기

```python
#set_style 메서드
sns.set_style('whitegrid')
fig, ax = plt.subplots()
ax = sns.violinplot(x='time', y='total_bill', hue='sex', data=tips, split=True)
```

```python
fig = plt.figure()

seaborn_styles = ['darkgrid', 'whitegrid', 'dark', 'white', 'ticks']

for idx, style in enumerate(seaborn_styles): #idx에 index, style에 리스트 값 반환
    plot_position = idx+1
    with sns.axes_style(style): #with로 자원 사용 후 반납
        ax = fig.add_subplot(2, 3, plot_position) #순서대로 배치
        violin = sns.violinplot(x='time', y='total_bill', data=tips, ax=ax)
        violin.set_title(style)

fig.tight_layout()
```
