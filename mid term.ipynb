```python
grades = [[2020101, "John Snow", "Database", 90],[2020102, "Jacky Chan", "Programming", 75]]

for x in range(len(grades)):
    student_id = grades[x][0]
    student_name = grades[x][1]
    course = grades[x][2]
    if grades[x][2] == "Database":
        instructor = "Prof. Gangnam"
    else:
        instructor = "Prof.Junggu"
    
    if grades[x][3] < 50:
        grade = "D"
    elif grades[x][3]<75:
        grade = "C"
    elif grades[x][3]<85:
        grade = "B"
    else:
        grade = "A"
    
    print("Student ID \t:", student_id)
    print("Student name \t:", student_name)
    print("Course name\t:", course )
    print("Instructor\t:", instructor)
    print("Grade Letter \t:", grade)
    print("--"*15)
    
```

```python
list_number=[3,6,8,9,2,3,4,5]
summ = 0
for x in range(len(list_number)):
    summ += list_number[x]
    
average = summ/len(list_number)

variance = 0
for x in range(len(list_number)):
    variance += (list_number[x]- average)**2

variance = variance/len(list_number)

standard_dev = variance**0.5

print(summ)
print(average)
print(variance)
print(standard_dev)
```

```python
A = [[3,7], [9,5]]
B = [[2,1], [4,8]]

for x in range(len(A)):
    for y in range(len(A[x])):
        print(A[x][y]+B[x][y])

for x in range(len(A)):
    for y in range(len(A[x])):
        c=0
        for z in range(len(B)):
            c+=A[x][z]*B[z][y]
        
        print(c)
```

```python
print("Welcome to the Course Manager v1.0")

while True:
    course_num = int(input("Please input the # of courses "))
    if course_num > 5:
        print("Sorry, maximum is 5 only!")
    else:
        break

course = open('course_list.txt','w')
course_list = []
for x in range(course_num):
    _input = input("Please type the course name: ")
    course_list.append(_input)
    _input = _input+"\n"
    course.write(_input)
course.close()
print("Course list : [", end="")
for x in range(course_num):
    print("\'",course_list[x],"\'",sep="",end="")
    if x!=(course_num-1):
        print(", ", end='')

print("]\nThe course list is saved in course_list.txt")
```
